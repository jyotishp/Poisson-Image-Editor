# Load Required Modules

```python
%matplotlib inline

import cv2
import matplotlib
import matplotlib.pyplot as plt
from importlib import reload
from poisson.core import Poisson
from poisson.utils import *
```

# Function to perform blending and display output

```python
def perform_blending(src, msk, target):
    plt.figure(figsize=(18,16), dpi=80, facecolor='w', edgecolor='k')
    plt.subplot(131)
    plt.imshow(cv2.cvtColor(cv2.imread(src), cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title('Source Image')

    plt.subplot(132)
    plt.imshow(cv2.cvtColor(cv2.imread(msk), cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title('Mask')

    plt.subplot(133)
    plt.imshow(cv2.cvtColor(cv2.imread(target), cv2.COLOR_BGR2RGB))
    plt.axis('off')
    _ = plt.title('Target Image')
    
    poisson_solver = Poisson(src, target, msk)
    poisson_solver.read_source(src)
    poisson_solver.read_target(target)
    poisson_solver.read_mask(msk)

    resultant = poisson_solver.seamless_blend()

    plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')

    plt.subplot(121)
    plt.imshow(resultant)
    plt.axis('off')
    plt.title('Seamless Blending')

    plt.subplot(122)
    plt.imshow(direct_patch(poisson_solver.source, poisson_solver.mask, poisson_solver.target))
    plt.axis('off')
    _ = plt.title('Direct Patching')
```
