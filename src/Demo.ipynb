# Load Required Modiles

```python
%matplotlib inline

import cv2
import matplotlib
import matplotlib.pyplot as plt
from poisson.core import Poisson
from poisson.utils import *
```

# Example 1
## Load the images

```python
source_path = '../test_data/1s.png'
mask_path = '../test_data/1m.png'
target_path = '../test_data/1t.png'

plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')

plt.subplot(131)
plt.imshow(cv2.cvtColor(cv2.imread(source_path), cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Source Image')

plt.subplot(132)
plt.imshow(cv2.cvtColor(cv2.imread(mask_path), cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Mask')

plt.subplot(133)
plt.imshow(cv2.cvtColor(cv2.imread(target_path), cv2.COLOR_BGR2RGB))
plt.axis('off')
_ = plt.title('Target Image')
```

## Perform Seamless Blending

```python
poisson_solver = Poisson(source_path, target_path, mask_path)
poisson_solver.read_source(source_path)
poisson_solver.read_target(target_path)
poisson_solver.read_mask(mask_path)

resultant = poisson_solver.seamless_blend()

plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')

plt.subplot(121)
plt.imshow(resultant)
plt.axis('off')
plt.title('Seamless Blending')

plt.subplot(122)
plt.imshow(direct_patch(poisson_solver.source, poisson_solver.mask, poisson_solver.target))
plt.axis('off')
_ = plt.title('Direct Patching')
```
